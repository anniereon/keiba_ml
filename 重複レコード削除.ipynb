```python
### インポート ###

import csv
import pandas as pd
from datetime import datetime
```

```python
### ファイル読み込み ###
file_path = "C:\\keiba\\tool\\history\\history.csv"
df = pd.read_csv(file_path, encoding="cp932")
```

```python
### キー重複するレコードを削除 ###
keys = ['race_id', 'horse_number']
print(df[df.duplicated(subset=keys, keep=False)])
df = df.drop_duplicates(subset=keys, keep='last')
print(df[df.duplicated(subset=keys, keep=False)])
```

```python
### CSV出力 ###
df.to_csv(file_path, index=False, encoding="cp932")
```
