```python
import numpy as np
import pandas as pd
import xgboost as xgb
from scipy import stats
import japanize_matplotlib
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
from tqdm import tqdm
import lightgbm as lgb
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
from sklearn.model_selection import KFold
import optuna

# ===============================================================
# é–¢æ•°å®šç¾©
# ===============================================================
def handle_missing_val(df, features, enabled, min_horses):
    """ãƒ¬ãƒ¼ã‚¹IDã”ã¨ã«ã‚°ãƒ«ãƒ¼ãƒ—åŒ–ã—ã€æ¬ æå‡¦ç†ã‚’è¡Œã†é–¢æ•°ã€‚"""
    if not enabled: return df
    
    processed_groups = []
    # race_id ã”ã¨ã«å‡¦ç†
    for race_id, group in df.groupby("race_id"):
        actual_min_horses = group["num_horses"].iloc[0]
        non_missing_mask = group[features].notnull().all(axis=1)
        
        if non_missing_mask.sum() < actual_min_horses: continue
            
        for col in features:
            if group[col].isnull().any():
                group[col] = group[col].fillna(group[col].mean(skipna=True))
        processed_groups.append(group)
        
    df_processed = pd.concat(processed_groups, ignore_index=True)
    return df_processed

def get_last_corner_position(row):
    """æœ€çµ‚ã‚³ãƒ¼ãƒŠãƒ¼é †ä½ã‚’å–å¾—ã™ã‚‹é–¢æ•°"""
    for i in [4, 3, 2, 1]:
        val = row.get(f"position_{i}")
        if pd.notnull(val): return val
    return np.nan

# ===============================================================
# ãƒ‡ãƒ¼ã‚¿èª­ã¿è¾¼ã¿ã¨åŸºæœ¬å‡¦ç†
# ===============================================================

target_num_horses = [12]
target_course_ids = [
    103, 201, 202, 302, 402,
    501, 502, 601, 602, 700, 702, 802,
    900, 901, 1001, 1002, 1201, 1304, 1400, 1402
]

input_file = "input_å…¨é ­_ç·¨é›†å¾Œ.csv"
df_all = pd.read_csv(input_file, encoding="utf-8")
df_all["race_date"] = pd.to_datetime(df_all["race_date"])
df_all = df_all.sort_values(["race_id", "horse_number"])

# ãƒ©ãƒ³ãƒ‹ãƒ³ã‚°ã‚¹ã‚¿ã‚¤ãƒ«åˆ¤å®š
df_all["last_corner_position"] = df_all.apply(get_last_corner_position, axis=1)
df_all["last_corner_position_rate"] = df_all["last_corner_position"] / df_all["num_horses"]
df_all = df_all.sort_values(["horse_id", "race_date"]).reset_index(drop=True)

def calc_past_avg(x): return x.expanding().mean().shift()
df_all["avg_last_corner_pos_rate_past"] = (
    df_all.groupby("horse_id", group_keys=False)["last_corner_position_rate"].transform(calc_past_avg)
)
df_all["running_style_type_past"] = np.where(
    df_all["avg_last_corner_pos_rate_past"] < 0.5, "front",
    np.where(df_all["avg_last_corner_pos_rate_past"].notnull(), "back", np.nan)
)

# å¯¾è±¡ã‚³ãƒ¼ã‚¹Ã—é ­æ•°ã«çµã‚‹
df_target = df_all[df_all["course_id"].isin(target_course_ids)].copy()
df_target = df_target[df_target["num_horses"].isin(target_num_horses)].copy()
missing_check_features = ["time_index_average_2", "jockey_place_rate_100"] 
df_target = handle_missing_val(df_target, features=missing_check_features, enabled=True, min_horses=min(target_num_horses))

# ===============================================================
# ç‰¹å¾´é‡ä½œæˆ (å…±é€š)
# ===============================================================

# 1. ç¾çŠ¶ãƒ»ç†æƒ³ã®ã‚¿ã‚¤ãƒ æŒ‡æ•°å·®
race_avg = df_target.groupby("race_id")["time_index_average_2"].transform("mean")
df_target["time_index_diff_from_avg"] = df_target["time_index_average_2"] - race_avg
race_avg_perfect = df_target.groupby("race_id")["time_index"].transform("mean")
df_target["time_index_diff_from_avg_perfect"] = df_target["time_index"] - race_avg_perfect

# 2. é¨æ‰‹å‹ç‡å·®
race_avg = df_target.groupby("race_id")["jockey_place_rate_100"].transform("mean")
df_target["jockey_place_rate_diff_from_avg"] = df_target["jockey_place_rate_100"] - race_avg

# 3. é¦¬ç•ªåˆ¥çµ±è¨ˆ (ä¿®æ­£æ¸ˆã¿)
df_target["is_win"] = (df_target["finish_rank"] == 1).astype(int)
df_filtered_for_win_rate = df_all[
    df_all["course_id"].isin(target_course_ids) & 
    df_all["num_horses"].isin(target_num_horses)
].copy()
df_filtered_for_win_rate["is_win"] = (df_filtered_for_win_rate["finish_rank"] == 1).astype(int)
win_rate_by_horse_number = (
    df_filtered_for_win_rate.groupby(["course_id", "horse_number"])["is_win"].mean()
    .reset_index().rename(columns={"is_win": "win_rate_by_course_horse_number"})
)
df_target = df_target.merge(win_rate_by_horse_number, on=["course_id", "horse_number"], how="left")
race_avg = df_target.groupby("race_id")["win_rate_by_course_horse_number"].transform("mean")
df_target["win_rate_by_course_horse_number_diff_from_avg"] = df_target["win_rate_by_course_horse_number"] - race_avg

# 4 & 5. ãƒ©ãƒ³ãƒ‹ãƒ³ã‚°ã‚¹ã‚¿ã‚¤ãƒ«ç‰¹å¾´é‡
# (4) ãƒ©ãƒ³ãƒ‹ãƒ³ã‚°ã‚¹ã‚¿ã‚¤ãƒ«åˆ¥å‹ç‡ï¼ˆã‚³ãƒ¼ã‚¹Ã—front/backï¼‰
horse_mean_style_rate = (
    df_all.groupby("horse_id")["last_corner_position_rate"].mean().reset_index()
    .rename(columns={"last_corner_position_rate": "mean_last_corner_pos_rate_all"})
)
horse_mean_style_rate["running_style_type_overall"] = np.where(
    horse_mean_style_rate["mean_last_corner_pos_rate_all"] < 0.5, "front", "back"
)
df_course = df_all[df_all["course_id"].isin(target_course_ids)].copy()
df_course = df_course.merge(horse_mean_style_rate[["horse_id", "running_style_type_overall"]], on="horse_id", how="left")
course_style_win_summary_target = (
    df_course.groupby(["course_id", "running_style_type_overall"])
    .agg(total_runs=("horse_id", "count"), total_wins=("finish_rank", lambda x: (x == 1).sum())).reset_index()
)
course_style_win_summary_target["win_rate_overall"] = course_style_win_summary_target["total_wins"] / course_style_win_summary_target["total_runs"]
df_target = df_target.merge(
    course_style_win_summary_target[["course_id", "running_style_type_overall", "win_rate_overall"]],
    left_on=["course_id", "running_style_type_past"], right_on=["course_id", "running_style_type_overall"], how="left"
)
df_target = df_target.drop(columns=["running_style_type_overall"]).rename(columns={"win_rate_overall": "running_style_win_prob"})
race_avg = df_target.groupby("race_id")["running_style_win_prob"].transform("mean")
df_target["running_style_win_prob_diff_from_avg"] = df_target["running_style_win_prob"] - race_avg

# (5) å‰æ–¹ï¼å¾Œæ–¹ã®å‰²åˆ Ã— å‹ç‡ (å‰å›ã®ã‚³ãƒ¼ãƒ‰ã§ä½œæˆæ¸ˆã¿ã¨ä»®å®š)
if "style_win_prob_by_ratio_diff_from_avg" not in df_target.columns:
    df_target["style_win_prob_by_ratio_diff_from_avg"] = 0 
    
# ===============================================================
# æ–½ç­–ï¼‘ï¼šæ‰‹å‹•è£œæ­£ (Manual Corrected) ã®ãŸã‚ã®ç‰¹å¾´é‡ä½œæˆ
# ===============================================================

# 1. å…¨ãƒ¬ãƒ¼ã‚¹ã®å¹³å‡ã‚¿ã‚¤ãƒ æŒ‡æ•°
overall_avg_time_index = df_target["time_index"].mean()

# 2. ã‚³ãƒ¼ã‚¹ã”ã¨ã®å¹³å‡ã‚¿ã‚¤ãƒ æŒ‡æ•°
course_avg_time_index = (
    df_target.groupby("course_id")["time_index"].mean()
    .reset_index().rename(columns={"time_index": "course_avg_time_index"})
)

df_target = df_target.merge(course_avg_time_index, on="course_id", how="left")

# 3. ã‚³ãƒ¼ã‚¹é–“å·® D ã®è¨ˆç®—
df_target["course_time_index_diff_D"] = df_target["course_avg_time_index"] - overall_avg_time_index

# 4. æ‰‹å‹•è£œæ­£ã•ã‚ŒãŸ time_index_average_2 ã®è¨ˆç®—
df_target["time_index_avg_2_corrected_manual"] = (
    df_target["time_index_average_2"] + df_target["course_time_index_diff_D"]
)

# 5. ç‰¹å¾´é‡åŒ– (ãƒ¬ãƒ¼ã‚¹å¹³å‡ã¨ã®å·®)
race_avg_manual = df_target.groupby("race_id")["time_index_avg_2_corrected_manual"].transform("mean")
df_target["time_index_diff_from_avg_manual"] = df_target["time_index_avg_2_corrected_manual"] - race_avg_manual

print(f"âœ… æ‰‹å‹•è£œæ­£ã«ã‚ˆã‚‹ç‰¹å¾´é‡ 'time_index_diff_from_avg_manual' ã‚’ä½œæˆã—ã¾ã—ãŸã€‚")

# ===============================================================
# LightGBMã«ã‚ˆã‚‹ã‚¿ã‚¤ãƒ æŒ‡æ•°äºˆæ¸¬ï¼ˆWalk-Forwardï¼‰
# ===============================================================

start_year = 2021
end_year = 2024
num_train_races = 100

# äºˆæ¸¬çµæœã‚’æ ¼ç´ã™ã‚‹è¾æ›¸
predicted_time_index_dfs = {}

# äºˆæ¸¬ã‚·ãƒŠãƒªã‚ªã¨ç›®çš„å¤‰æ•°ã®è¨­å®š
lgbm_scenarios = {
    "LGBM_Predicted": "time_index", # é€šå¸¸äºˆæ¸¬
    "LGBM_Residual": "time_index_residual", # æ®‹å·®äºˆæ¸¬ (time_index - time_index_average_2)
    "LGBM_TunedResidual": "time_index_residual", # ãƒãƒ¥ãƒ¼ãƒ‹ãƒ³ã‚°æ¸ˆã¿æ®‹å·®äºˆæ¸¬
}

# ç›®çš„å¤‰æ•°ã®ä½œæˆ
df_target["time_index_residual"] = df_target["time_index"] - df_target["time_index_average_2"]

# --- ã€ä¿®æ­£æ¸ˆã¿ã€‘LGBMãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã®çµ±ä¸€ ---
# å…¨ã‚·ãƒŠãƒªã‚ªã§å…±é€šã®ãƒãƒ¥ãƒ¼ãƒ‹ãƒ³ã‚°æ¸ˆã¿ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã‚’å®šç¾© (max_depth=3ãªã©)
common_lgbm_params = {
    'objective': 'regression', 
    'metric': 'rmse', 
    'n_estimators': 500, 
    'learning_rate': 0.03, 
    'verbose': -1, 
    'n_jobs': -1, 
    'seed': 42,
    'max_depth': 3,
    'num_leaves': 50, 
    'reg_alpha': 0.8, 
    'reg_lambda': 0.5, 
    'min_child_samples': 20
}

# å„ã‚·ãƒŠãƒªã‚ªã«å…±é€šãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã‚’é©ç”¨
lgbm_params = {
    "LGBM_Predicted": common_lgbm_params,
    "LGBM_Residual": common_lgbm_params,
    "LGBM_TunedResidual": common_lgbm_params
}
# ----------------------------------------

lgbm_features = {
    "LGBM_Predicted": ["time_index_by_course_horse_number", "time_index_average_2", "jockey_place_rate_diff_from_avg"],
    "LGBM_Residual": ["time_index_by_course_horse_number", "time_index_average_2", "jockey_place_rate_diff_from_avg"],
    # TunedResidual ã®ã¿ã‚³ãƒ¼ã‚¹ç‰¹æ€§ã‚’å­¦ç¿’ã•ã›ã‚‹ãŸã‚ç‰¹å¾´é‡ãƒªã‚¹ãƒˆãŒç•°ãªã‚‹
    "LGBM_TunedResidual": [
        "time_index_by_course_horse_number", "time_index_average_2", "jockey_place_rate_diff_from_avg", 
        "course_id", "num_horses" 
    ]
}

# (1) ã‚³ãƒ¼ã‚¹Ã—é¦¬ç•ªã”ã¨ã®ã‚¿ã‚¤ãƒ æŒ‡æ•°å¹³å‡ (time_index) ã‚’ç®—å‡º (LGBMç‰¹å¾´é‡)
df_filtered_for_time_rate = df_all[
    df_all["course_id"].isin(target_course_ids) & df_all["num_horses"].isin(target_num_horses)
].copy()
time_index_by_course_horse_number = (
    df_filtered_for_time_rate.groupby(["course_id", "horse_number"])["time_index"].mean()
    .reset_index().rename(columns={"time_index": "time_index_by_course_horse_number"})
)
df_target = df_target.merge(time_index_by_course_horse_number, on=["course_id", "horse_number"], how="left")
df_target["time_index_by_course_horse_number"] = df_target["time_index_by_course_horse_number"].fillna(df_target["time_index_by_course_horse_number"].mean())


# Walk-Forwardãƒ«ãƒ¼ãƒ— (LGBM)
for scenario_name, target_var in lgbm_scenarios.items():
    print(f"\n" + "="*50)
    print(f"ğŸ“ˆ ã‚¿ã‚¤ãƒ æŒ‡æ•°äºˆæ¸¬å®Ÿè¡Œ: {scenario_name}")
    print("="*50)
    
    wf_predictions_list = []
    features_list = lgbm_features[scenario_name]
    lgbm_param = lgbm_params[scenario_name]

    for course_id in target_course_ids:
        df_course = df_target[
            (df_target["course_id"] == course_id) &
            (df_target["race_date"].dt.year >= start_year) &
            (df_target["race_date"].dt.year <= end_year)
        ].copy()

        if df_course.empty: continue
        
        df_course = df_course.sort_values(["race_date", "race_id", "horse_number"]).reset_index(drop=True)
        race_ids = df_course["race_id"].unique()

        if len(race_ids) <= num_train_races: continue

        for i in tqdm(range(num_train_races, len(race_ids)), desc=f"WF ({scenario_name} | {course_id})"):

            train_races = race_ids[:i]
            val_race = race_ids[i]

            df_train = df_course[df_course["race_id"].isin(train_races)].copy()
            df_val = df_course[df_course["race_id"] == val_race].copy()

            df_train.dropna(subset=features_list + [target_var], inplace=True)
            
            # ã€ä¿®æ­£ç®‡æ‰€ã€‘ .copy() ã‚’è¿½åŠ ã—ã¦ SettingWithCopyWarning ã‚’å›é¿
            X_train = df_train[features_list].copy()
            y_train = df_train[target_var]
            X_val = df_val[features_list].copy()
            
            time_model = lgb.LGBMRegressor(**lgbm_param)
            
            # ã‚«ãƒ†ã‚´ãƒªå¤‰æ•°ã®æŒ‡å®š (TunedResidual ã®ã¿ course_id ã‚’å«ã‚€)
            cat_features = ["course_id"] if "course_id" in features_list else None
            
            try:
                # course_id ã‚’ã‚«ãƒ†ã‚´ãƒªã‚«ãƒ«ç‰¹å¾´é‡ã¨ã—ã¦æ‰±ã†ãŸã‚ã«ãƒ‡ãƒ¼ã‚¿å‹ã‚’å¤‰æ›
                if "course_id" in features_list:
                    # è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã¨æ¤œè¨¼ãƒ‡ãƒ¼ã‚¿ã®ã‚³ãƒ¼ã‚¹IDã‚’ 'category' å‹ã«å¤‰æ›
                    X_train["course_id"] = X_train["course_id"].astype("category")
                    if "course_id" in X_val.columns:
                        X_val["course_id"] = X_val["course_id"].astype("category")

                time_model.fit(X_train, y_train, categorical_feature=cat_features)
            except lgb.basic.LightGBMError:
                # ã‚«ãƒ†ã‚´ãƒªã‚«ãƒ«ç‰¹å¾´é‡ã®æŒ‡å®šã§ã‚¨ãƒ©ãƒ¼ãŒå‡ºãŸå ´åˆã€æŒ‡å®šãªã—ã§å†è©¦è¡Œ
                time_model.fit(X_train, y_train)

            df_val["predicted_val"] = time_model.predict(X_val)
            
            if scenario_name.endswith("Residual"):
                # æ®‹å·®äºˆæ¸¬ã®å ´åˆã€predicted_val ã¯è£œæ­£é …ãªã®ã§ time_index_average_2 ã«è¶³ã™
                df_val["predicted_time_index"] = df_val["time_index_average_2"] + df_val["predicted_val"]
            else:
                df_val["predicted_time_index"] = df_val["predicted_val"]
            
            # --- äºˆæ¸¬çµæœã®ä¸€æ™‚ãƒªã‚¹ãƒˆã¸ã®æ ¼ç´ ---
            wf_predictions_list.append(df_val)

    # --- Walk-Forwardãƒ«ãƒ¼ãƒ—çµ‚äº†å¾Œã®å‡¦ç† ---
    df_predicted_results = pd.concat(wf_predictions_list, ignore_index=True)
    
    # äºˆæ¸¬çµæœã‚’æ ¼ç´ã™ã‚‹æœ€çµ‚ã‚«ãƒ©ãƒ å
    final_predicted_col = f"predicted_time_index_{scenario_name}"
    
    # df_targetã«è©²å½“ã‚«ãƒ©ãƒ ãŒå­˜åœ¨ã—ãªã„å ´åˆã¯NaNã§åˆæœŸåŒ–
    if final_predicted_col not in df_target.columns:
        df_target[final_predicted_col] = np.nan
        
    # äºˆæ¸¬çµæœã®ãƒãƒ¼ã‚¸
    merge_df = df_predicted_results[["race_id", "horse_number", "predicted_time_index"]].rename(
        columns={"predicted_time_index": final_predicted_col}
    )
    
    temp_df = df_target[["race_id", "horse_number"]].merge(
        merge_df,
        on=["race_id", "horse_number"],
        how="left"
    )
    
    df_target[final_predicted_col] = temp_df[final_predicted_col].values
    
    # äºˆæ¸¬å€¤ã‚’ä½¿ã£ã¦ç‰¹å¾´é‡ã‚’ä½œæˆ
    race_avg_predicted = df_target.groupby("race_id")[final_predicted_col].transform("mean")
    df_target[f"time_index_diff_from_avg_{scenario_name}"] = df_target[final_predicted_col] - race_avg_predicted


print(f"âœ… LightGBMã«ã‚ˆã‚‹å…¨äºˆæ¸¬ã‚·ãƒŠãƒªã‚ªã®ç‰¹å¾´é‡ã‚’df_targetã«ãƒãƒ¼ã‚¸ã—ã¾ã—ãŸã€‚")

# ===============================================================
# æœ€çµ‚ãƒ¢ãƒ‡ãƒ« (XGBoost) ã®ã‚·ãƒŠãƒªã‚ªè¨­å®š
# ===============================================================

# ã€ä¿®æ­£æ¸ˆã¿ã€‘features_baseã®ã‚³ãƒ¡ãƒ³ãƒˆã‚¢ã‚¦ãƒˆã‚’è§£é™¤ (å…¨ã‚·ãƒŠãƒªã‚ªã§å…±é€šç‰¹å¾´é‡ã‚’ä½¿ç”¨)
features_base = [
#    "jockey_place_rate_diff_from_avg",
#    "win_rate_by_course_horse_number_diff_from_avg",
#    "running_style_win_prob_diff_from_avg",
#    "style_win_prob_by_ratio_diff_from_avg",
#    "num_horses",
]

scenarios_main = {
    "Model (ç¾çŠ¶)": ["time_index_diff_from_avg"] + features_base,
    "Predicted (LGBMäºˆæ¸¬)": ["time_index_diff_from_avg_LGBM_Predicted"] + features_base,
    "Manual Corrected (æ‰‹å‹•è£œæ­£)": ["time_index_diff_from_avg_manual"] + features_base,
    "LGBM Residual (æ®‹å·®äºˆæ¸¬)": ["time_index_diff_from_avg_LGBM_Residual"] + features_base,
}

# äºˆæ¸¬çµæœã‚’æ ¼ç´ã™ã‚‹è¾æ›¸ã‚’åˆæœŸåŒ–
all_wf_predictions = {name: [] for name in scenarios_main.keys()}
df_target["relative_rank"] = df_target["finish_rank"] / df_target["num_horses"]

# ===============================================================
# ğŸ” æœ€çµ‚ãƒ¢ãƒ‡ãƒ«ã® Walk-Forward (XGBoost) - 5ã‚·ãƒŠãƒªã‚ªå®Ÿè¡Œ
# ===============================================================

print("\n\n" + "#"*70)
print("ğŸ ã‚¹ãƒ†ãƒƒãƒ—3ï¼šæœ€çµ‚ XGBoost ãƒ¢ãƒ‡ãƒ«ã«ã‚ˆã‚‹5ã‚·ãƒŠãƒªã‚ªæ¯”è¼ƒ")
print("#"*70)

# XGBoostã®ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ (max_depth=3ã¯çµ±ä¸€ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã¨ã—ã¦ç¶­æŒ)
xgb_params = {
    "objective": "binary:logistic", 
    "eval_metric": "logloss", 
    "random_state": 42,
    "subsample": 0.8, 
    "colsample_bytree": 0.8, 
    "base_score": 0.5, 
    "max_depth": 3,
    "learning_rate": 0.06, 
    "n_estimators": 100, 
    "reg_alpha": 0.5, 
    "reg_lambda": 1.0,
}

for course_id in target_course_ids:

    df_course = df_target[
        (df_target["course_id"] == course_id) &
        (df_target["race_date"].dt.year >= start_year) &
        (df_target["race_date"].dt.year <= end_year)
    ].copy()

    # LGBMäºˆæ¸¬å€¤ï¼ˆPredictedã¨Residual/TunedResidualï¼‰ã‚’æŒã¤ãƒ¬ãƒ¼ã‚¹ã®ã¿ã‚’ä½¿ç”¨
    df_course = df_course[
        df_course["predicted_time_index_LGBM_Predicted"].notnull()
    ].copy()

    if df_course.empty: continue
    
    df_course = df_course.sort_values(["race_date", "race_id", "horse_number"]).reset_index(drop=True)
    race_ids = df_course["race_id"].unique()

    if len(race_ids) <= num_train_races: continue

    print(f"\nâ–¶ ã‚³ãƒ¼ã‚¹ {course_id} å¯¾è±¡ãƒ¬ãƒ¼ã‚¹æ•°: {len(race_ids)}")

    for scenario_name, features_list in scenarios_main.items():
        wf_predictions_list = []
        
        for i in tqdm(range(num_train_races, len(race_ids)),
                      desc=f"XGB WF ({course_id} | {scenario_name})"):

            train_races = race_ids[:i]
            val_race = race_ids[i]

            df_train = df_course[df_course["race_id"].isin(train_races)].copy()
            df_val = df_course[df_course["race_id"] == val_race].copy()

            # äººæ°—1ç•ªã‚’é™¤å¤–ã™ã‚‹ãƒ•ã‚£ãƒ«ã‚¿ãƒªãƒ³ã‚°
            df_train_filtered = df_train[~df_train["popularity"].isin([1])].dropna(subset=features_list)

            X_train = df_train_filtered[features_list]
            y_train = (df_train_filtered["relative_rank"] <= 0.5).astype(int) 
            # æ¤œè¨¼ãƒ‡ãƒ¼ã‚¿ã¯NaNã‚’0ã§è£œå®Œï¼ˆå­¦ç¿’ãƒ‡ãƒ¼ã‚¿ã§NaNæ’é™¤æ¸ˆã¿ï¼‰
            X_val = df_val[features_list].fillna(0) 

            model_wf = XGBClassifier(**xgb_params)
            model_wf.fit(X_train, y_train)

            df_val["pred_prob"] = model_wf.predict_proba(X_val)[:, 1]
            df_val = df_val.sort_values("pred_prob", ascending=False)
            
            wf_predictions_list.append(df_val)

        wf_pred_df = pd.concat(wf_predictions_list, ignore_index=True)
        all_wf_predictions[scenario_name].append(wf_pred_df)

# ===============================================================
# ğŸ”¹ æœ€çµ‚çµæœã¾ã¨ã‚ï¼ˆå„ãƒ¬ãƒ¼ã‚¹ 1é ­é¸æŠï¼‰
# ===============================================================
print("\n\n" + "="*70)
print("ğŸ’° æœ€çµ‚çµæœï¼š5ã‚·ãƒŠãƒªã‚ªã®å›åç‡ãƒ»çš„ä¸­ç‡æ¯”è¼ƒ")
print("="*70)

summary_results = []
bet_amount = 100

for scenario_name, list_of_dfs in all_wf_predictions.items():
    if not list_of_dfs: continue

    df_pred_all_courses = pd.concat(list_of_dfs, ignore_index=True)
    df_pred_top1 = df_pred_all_courses.loc[
        df_pred_all_courses.groupby("race_id")["pred_prob"].idxmax()
    ].copy()

    total_bet = len(df_pred_top1) * bet_amount
    
    # å˜å‹è©•ä¾¡
    df_pred_top1["is_win_hit"] = (df_pred_top1["finish_rank"] == 1).astype(int)
    df_pred_top1["payout_win"] = df_pred_top1["is_win_hit"] * df_pred_top1["odds"] * bet_amount
    total_payout_win = df_pred_top1["payout_win"].sum()
    
    win_accuracy = df_pred_top1["is_win_hit"].mean() * 100
    roi_win = total_payout_win / total_bet * 100 if total_bet > 0 else 0
    
    # è¤‡å‹è©•ä¾¡
    df_pred_top1["is_place_hit"] = (df_pred_top1["finish_rank"] <= 3).astype(int) 
    place_accuracy = df_pred_top1["is_place_hit"].mean() * 100

    summary_results.append({
        "Scenario": scenario_name,
        "Total Races": len(df_pred_top1),
        "Win Accuracy (%)" : win_accuracy,
        "Place Accuracy (%)" : place_accuracy,
        "ROI (å˜å‹å›åç‡ %)" : roi_win,
    })

df_summary = pd.DataFrame(summary_results)
print(df_summary.to_markdown(index=False, floatfmt=".2f"))

# ===============================================================
# ğŸ“Š çµæœã®å¯è¦–åŒ–
# ===============================================================

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

colors = ['gray', 'skyblue', 'lightgreen', 'orange', 'red'] 

# 1. å›åç‡ã®æ¯”è¼ƒ (å˜å‹äºˆæƒ³)
axes[0].bar(df_summary["Scenario"], df_summary["ROI (å˜å‹å›åç‡ %)"], color=colors)
axes[0].axhline(100, color='r', linestyle='--', linewidth=1, label='ROI 100% (å…ƒæœ¬å›å)')
axes[0].set_title("å˜å‹äºˆæƒ³ã®å›åç‡æ¯”è¼ƒ (5ã‚·ãƒŠãƒªã‚ª)", fontsize=14)
axes[0].set_ylabel("å›åç‡ (%)", fontsize=12)
axes[0].set_xlabel("ã‚·ãƒŠãƒªã‚ª", fontsize=12)
axes[0].tick_params(axis='x', rotation=15)
axes[0].legend()
axes[0].grid(axis='y', linestyle='--')
for i, v in enumerate(df_summary["ROI (å˜å‹å›åç‡ %)"]):
    axes[0].text(i, v + 1, f"{v:.2f}%", ha='center', va='bottom', fontweight='bold')

# 2. çš„ä¸­ç‡ã®æ¯”è¼ƒ (å˜å‹ãƒ»è¤‡å‹)
x = np.arange(len(df_summary["Scenario"]))  
width = 0.35  

rects1 = axes[1].bar(x - width/2, df_summary["Win Accuracy (%)"], width, label='å˜å‹çš„ä¸­ç‡', color=colors)
rects2 = axes[1].bar(x + width/2, df_summary["Place Accuracy (%)"], width, label='è¤‡å‹çš„ä¸­ç‡', color=['#A9D0F5', '#A9D0F5', '#A9D0F5', '#A9D0F5', '#A9D0F5'])

axes[1].set_title("çš„ä¸­ç‡ã®æ¯”è¼ƒ (5ã‚·ãƒŠãƒªã‚ª)", fontsize=14)
axes[1].set_ylabel("çš„ä¸­ç‡ (%)", fontsize=12)
axes[1].set_xlabel("ã‚·ãƒŠãƒªã‚ª", fontsize=12)
axes[1].set_xticks(x, df_summary["Scenario"], rotation=15)
axes[1].legend(loc='upper left')
axes[1].grid(axis='y', linestyle='--')

# ãƒ©ãƒ™ãƒ«è¿½åŠ 
def autolabel(rects, data):
    for rect, val in zip(rects, data):
        height = rect.get_height()
        axes[1].annotate(f'{val:.2f}%',
                         xy=(rect.get_x() + rect.get_width() / 2, height),
                         xytext=(0, 3), 
                         textcoords="offset points",
                         ha='center', va='bottom', fontsize=9)

autolabel(rects1, df_summary["Win Accuracy (%)"])
autolabel(rects2, df_summary["Place Accuracy (%)"])

plt.tight_layout()
plt.show() # 

print("\n\n" + "="*70)
print("ğŸ“Š è¿½åŠ åˆ†æï¼šã‚¿ã‚¤ãƒ æŒ‡æ•°äºˆæ¸¬èª¤å·®ã®æ¯”è¼ƒï¼ˆLGBMäºˆæ¸¬ vs æ‰‹å‹•è£œæ­£ï¼‰")
print("="*70)

# äºˆæ¸¬å€¤ãŒã‚ã‚‹ãƒ¬ãƒ¼ã‚¹ã«é™å®š
df_comparison = df_target[df_target["time_index_diff_from_avg_LGBM_Predicted"].notnull()].copy()

if not df_comparison.empty:
    # èª¤å·®ã®çµ¶å¯¾å€¤
    df_comparison["error_avg_2"] = np.abs(df_comparison["time_index_average_2"] - df_comparison["time_index"])
    df_comparison["error_manual"] = np.abs(df_comparison["time_index_avg_2_corrected_manual"] - df_comparison["time_index"])
    df_comparison["error_residual"] = np.abs(df_comparison["predicted_time_index_LGBM_Residual"] - df_comparison["time_index"])
    df_comparison["error_tuned_residual"] = np.abs(df_comparison["predicted_time_index_LGBM_TunedResidual"] - df_comparison["time_index"])

    # è©•ä¾¡æŒ‡æ¨™ã®è¨ˆç®— (MAE)
    mae_data = {
        "time_index_average_2 (éå»å®Ÿç¸¾å¹³å‡)": df_comparison["error_avg_2"].mean(),
        "Manual Corrected (æ‰‹å‹•è£œæ­£)": df_comparison["error_manual"].mean(),
        "LGBM Residual (æ®‹å·®äºˆæ¸¬)": df_comparison["error_residual"].mean(),
    }

    mae_df = pd.DataFrame({
        "ãƒ¢ãƒ‡ãƒ«": mae_data.keys(),
        "MAE (å¹³å‡çµ¶å¯¾èª¤å·®)": mae_data.values()
    }).sort_values(by="MAE (å¹³å‡çµ¶å¯¾èª¤å·®)")
    
    print("\n## å®Ÿç¸¾ã‚¿ã‚¤ãƒ æŒ‡æ•° (time_index) ã«å¯¾ã™ã‚‹äºˆæ¸¬èª¤å·® (MAE) ã®æ¯”è¼ƒ")
    print(mae_df.to_markdown(index=False, floatfmt=".4f"))

else:
    print("âš  äºˆæ¸¬å€¤ã‚’æŒã¤ãƒ¬ãƒ¼ã‚¹ãƒ‡ãƒ¼ã‚¿ãŒå­˜åœ¨ã—ã¾ã›ã‚“ã€‚Walk-ForwardãŒæ­£ã—ãå®Ÿè¡Œã•ã‚ŒãŸã‹ç¢ºèªã—ã¦ãã ã•ã„ã€‚")
```
