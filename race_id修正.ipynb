```python
### インポート ###

import csv
import pandas as pd
from datetime import datetime
```

```python
### 競馬場の場所コードをキーにした辞書 ###
dic_racecourse = {
    30: "門別",
    35: "盛岡",
    36: "水沢",
    42: "浦和",
    43: "船橋",
    44: "大井",
    45: "川崎",
    46: "金沢",
    47: "笠松",
    48: "名古屋",
    50: "園田",
    54: "高知",
    55: "佐賀",
    65: "帯広(ば)",
}
dic_racecourse_id = {v: k for k, v in dic_racecourse.items()}
```

```python
####################
# ファイル読み込み
####################

# file_path = "C:\\keiba\\tool\\history\\history.csv"
file_path = "C:\\keiba\\tool\\train\\train_raw.csv"
df = pd.read_csv(file_path, encoding="cp932")

####################
# race_id修正
####################

# 年
df['yyyy'] = df['race_date'].str[:4]
# 競馬場ID
df["racecourse_id"] = df["racecourse"].map(dic_racecourse_id).fillna(0).astype(int).astype(str)
# 月日
df['mmdd'] = pd.to_datetime(df['race_date']).dt.strftime('%m%d')
# R番号
df['race_number_2digit'] = df['race_number'].astype(str).str.replace('R', '', regex=False).str.zfill(2)
# race_id
# すべてを文字列型に固定してから、catで連結する
df["race_id"] = df['yyyy'].astype(str) + df['racecourse_id'].astype(str) + df['mmdd'].astype(str) + df['race_number_2digit'].astype(str)
# 不要な列削除
df.drop(columns=['yyyy', 'racecourse_id', 'mmdd', 'race_number_2digit'], inplace=True)

############
# CSV出力
############

df.to_csv(file_path, index=False, encoding="cp932")
```
