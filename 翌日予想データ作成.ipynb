```python
import pandas as pd
```

```python
# CSVファイルを読み込む
df_8_learn = pd.read_csv('input_8頭_編集後.csv', encoding='utf-8')
df_9_learn = pd.read_csv('input_9頭_編集後.csv', encoding='utf-8')
df_10_learn = pd.read_csv('input_10頭_編集後.csv', encoding='utf-8')
df_11_learn = pd.read_csv('input_11頭_編集後.csv', encoding='utf-8')
df_12_learn = pd.read_csv('input_12頭_編集後.csv', encoding='utf-8')
df_all_learn = pd.read_csv('input_all_編集後.csv', encoding='utf-8')
df_pred = pd.read_csv('race_data.csv', encoding='cp932')
```

```python
# race_predに対して説明変数を追加する（タイム指数平均、騎手勝率）

# 予想上位確率を計算する

# 以降の処理は別ファイル

# df_pred読み込み

# 定時処理

# 時刻になったらスクレイピングしてオッズを取得

# オッズと予想上位確率が閾値の範囲内であれば購入（spat4で購入）

# 結果をメールで飛ばす

# あたったかどうかを確認し結果をメールで飛ばす
```

```python
# 過去レースのタイム指数の平均を計算して返す
def compute_time_index_avg(df_learn, df_pred):
    result = []

    for idx, row in df_pred.iterrows():
        horse_id = row['horse_id']
        race_date = row['race_date']

        past_races = df_learn[
            (df_learn['horse_id'] == horse_id) &
            (df_learn['race_date'] < race_date)
        ].sort_values(by='race_date', ascending=False).head(2)

        if len(past_races) >= 1:
            avg_time_index = past_races['time_index'].mean()
        else:
            avg_time_index = None  # 欠損として処理

        result.append(avg_time_index)

    return result
```

```python
# 騎手の過去100レースの勝率を計算して返す
def compute_jockey_place_rate(df_learn, df_pred):
    for idx, row in df_pred.iterrows():
        jockey_id = row['jockey_id']
        race_date = row['race_date']
        
        past_races = df_learn[
            (df_learn['jockey_id'] == jockey_id) &
            (df_learn['race_date'] < race_date)
        ].sort_values(by='race_date', ascending=False).head(100)

        if not past_races.empty:
            wins = (past_races['finish_rank'] == 1).sum()
            win_rate = wins / len(past_races)
        else:
            win_rate = None
        
        # ✅ ここで直接書き込むのが正解！
        df_pred.at[idx, 'jockey_place_rate'] = win_rate

    return df_pred
```

```python
# race_predに対して説明変数を追加する（タイム指数平均、騎手勝率）

# １．日付をdatetime型に変換
df_all_learn['race_date'] = pd.to_datetime(df_all_learn['race_date'])
df_pred['race_date'] = pd.to_datetime(df_pred['race_date'])
# ２．jockey_idを文字列に変換して0埋めして5桁に揃える
df_all_learn['jockey_id'] = df_all_learn['jockey_id'].astype(str).str.zfill(5)
df_pred['jockey_id'] = df_pred['jockey_id'].astype(str).str.zfill(5)
# ３．time_index_avg の計算
df_pred['time_index_avg'] = compute_time_index_avg(df_all_learn, df_pred)
# ４．jockey_place_rate の計算
df_pred = compute_jockey_place_rate(df_all_learn, df_pred)
# ５．結果の保存
df_pred.to_csv('df_pred.csv', index=False, encoding='cp932')
print("特徴量を追加したファイルを保存しました：df_pred.csv")
```
